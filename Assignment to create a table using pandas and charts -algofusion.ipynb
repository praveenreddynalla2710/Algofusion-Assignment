```python
#year wise average score
year_wise_average=df.groupby('year')['marks'].mean()
year_wise_average
```

```python
#semester wise average score
semester_wise_average=df.groupby('semester')['marks'].mean()
semester_wise_average
```

```python
#subdivision wise average score
subdivision_average=df.groupby('subdivisions')['marks'].mean()
subdivision_average
```

```python
#subject scores grade wise
subject_scores_grade=df.groupby('grade')['marks'].mean()
subject_scores_grade
```

```python
#categorizing of marks
def performance (marks):
    if marks>=90:
        return 'excellent'
    elif marks>=80:
        return 'very good'
    elif marks>=70:
        return 'good'
    elif marks>=60:
        return 'better'
    elif marks>=50:
        return 'average'
    elif marks>=40:
        return 'Pass'
    else:
        return 'Fail'
```

```python
#calling the function
df['performance']=df['marks'].apply(performance)
```
